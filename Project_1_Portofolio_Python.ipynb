```python
from apyori import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#menampilkan ukuran data
data_transaksi.shape
```

```python
#hapus kolom transaksi
df1 = data_transaksi.iloc[:,1:]
df1
```

```python
result_df = pd.DataFrame(columns=['Kolom', 'Kategori', 'Frekuensi'])

for column in df1.columns:
    value_counts = df1[column].value_counts().reset_index()
    value_counts.columns = ['Kategori', 'Frekuensi']
    value_counts['Kolom'] = column
    result_df = pd.concat([result_df, value_counts], ignore_index=True)

print(result_df)
```

```python
# Menghapus baris dengan Kategori 0
result_df1 = result_df.query('Kategori == 1')
print(result_df1)
```

```python
result_df1 = result_df1.sort_values(by='Frekuensi', ascending=True)

plt.figure(figsize=(12, 8))
plt.barh(result_df1 ['Kolom'], width=result_df1['Frekuensi'], color='skyblue')
plt.xlabel('Frekuensi')
plt.ylabel('Item')
plt.title('Frekuensi Item Set')
plt.show()
```

```python
frequent_itemsets1 = apriori(df1, use_colnames = True, min_support = 0.3)
frequent_itemsets1
```

```python
rules1 = association_rules(frequent_itemsets1, metric = "confidence", min_threshold = 0.5)
rules1
```

```python
rules1.nlargest(n = 10, columns = [ 'lift','confidence','support'])
```
